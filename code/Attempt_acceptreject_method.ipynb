```python
import numpy as np
import matplotlib.pyplot as plt
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def distribution(e, n):
    return ( e+1) * (1 - n)**e

e = 10
N = np.linspace(0, 1, 200)
P = distribution(e, N)

plt.plot(N, P)
plt.xlabel('n')
plt.ylabel('p(n)')
plt.title('Distribution p(n) = (ζ+1)(1 - n)^ζ')
plt.grid(True)
plt.show()
```

```python


def target_distribution(e, n):
    return (e+1) * (1 - n)**e

def proposal_distribution():
    return random.uniform(0, 1)  # Uniform distribution between 0 and 1

def rejection_method(e, c):
    while True:
        # Step 1: Simulate the value of Y using the proposal distribution q(n)
        Y = proposal_distribution()
        
        # Step 2: Generate a random number from U
        U = random.uniform(0, 1)
        
        # Step 3: Accept or reject the value of Y
        if U <= target_distribution(e, Y) / (c * proposal_distribution()):
            # Accept Y and stop
            X = Y
            return X

e = 10
c = 5 # Choose an appropriate value for c
num_samples = 2000

samples = []
for _ in range(num_samples):
    sample = rejection_method(e, c)
    samples.append(sample)

# Plotting the histogram of the samples
plt.hist(samples, bins=20, density=True, alpha=0.5)
plt.xlabel('n')
plt.ylabel('Probability')
plt.title('Samples from the target distribution')
plt.show()
```
