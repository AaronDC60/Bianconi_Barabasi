### Visualizing distributions

Visualizing all the different distributions will allow us to check if the generator samples from the right distribution and give us insight into how the distribution looks like. 

#### Initialization

```python
# Import the fitness file containing the sample generator
import sys
sys.path.insert(0, '../')

from src import fitness

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

```python
# Initialize a fitness value generator
generator = fitness.generator()

# Number of samples to generate
n_samples = 10000
```

```python
type(generator)
```

#### Uniform distribution

```python
generator.set_current_distribution('uniform')

# Generate values
y_samples = []

for _ in range(n_samples):
    y_samples.append(generator.generate_value())
```

```python
# Plot the distribution
fig, axs = plt.subplots(1,2, figsize=(21, 7), sharex=True, sharey=True)
axs[0].hist(y_samples, density=True, bins = 50, width=0.015, color = 'black')
axs[0].set_title('Sampled uniform distribution', fontsize = 20)
axs[0].set_ylabel('$p_x$', fontsize=18)

axs[1].plot(np.linspace(0, 1, 1000), [1 for _ in range(1000)], color = 'black')
axs[1].set_title('Analytical uniform distribution', fontsize = 20)
for ax in axs:
    ax.set_xlim(0, 1)
    ax.set_xlabel('$x$', fontsize=18)

plt.tight_layout()
plt.show()
```

#### Exponential distribution ($\lambda$ = 1)

```python
generator.set_current_distribution('exponential')

# Generate values
y_samples = []

for _ in range(n_samples):
    y_samples.append(generator.generate_value())
```

```python
# Plot the distribution
fig, axs = plt.subplots(1,2, figsize=(21, 7), sharex=True, sharey=True)
_, bins,_ = axs[0].hist(y_samples, density=True, bins = 50, width=0.1, color = 'black')
axs[0].set_title('Sampled exponential distribution', fontsize = 20)
axs[0].set_ylabel('$p_x$', fontsize=18)

axs[1].plot(np.linspace(0, max(bins), 1000), [np.exp(-x) for x in np.linspace(0, max(bins), 1000)], color = 'black')
axs[1].set_title('Analytical exponential distribution', fontsize = 20)
for ax in axs:
    ax.set_xlim(left=0)
    ax.set_xlabel('$x$', fontsize=18)

plt.tight_layout()
plt.show()
```

#### Poisson distribution ($\lambda$ = 4)

```python
generator.set_current_distribution('poisson')
generator.set_param_poisson(4)

# Generate values
y_samples = []

for _ in range(n_samples):
    y_samples.append(generator.generate_value())
```

```python
# Plot the distribution
fig, axs = plt.subplots(1,2, figsize=(21, 7), sharex=True, sharey=True)
_, bins,_ = axs[0].hist(y_samples, density=True, bins = 15, width=0.5, color = 'black')
axs[0].set_title('Sampled poisson distribution', fontsize = 20)
axs[0].set_ylabel('$p_x$', fontsize=18)

axs[1].plot(np.arange(0, np.round(max(bins)) + 1, 1), [(4**x * np.exp(-4))/(np.math.factorial(int(x))) for x in np.arange(0, np.round(max(bins)) + 1, 1)], color = 'black')
axs[1].set_title('Analytical poisson distribution', fontsize = 20)
for ax in axs:
    ax.set_ylim(top=.5)
    ax.set_xlim(left=0)
    ax.set_xlabel('$x$', fontsize=18)

plt.tight_layout()
plt.show()
```

#### $\beta$-distribution ($\alpha = 2$, $\beta$ = 2)

```python
generator.set_current_distribution('beta')
generator.set_param_beta(2, 2)

# Generate values
y_samples = []

for _ in range(n_samples):
    y_samples.append(generator.generate_value())
```

```python
# Plot the distribution
fig, axs = plt.subplots(1,2, figsize=(21, 7), sharex=True, sharey=True)
_, bins,_ = axs[0].hist(y_samples, density=True, bins = 50, width=0.1, color = 'black')
axs[0].set_title('Sampled beta distribution', fontsize = 20)
axs[0].set_ylabel('$p_x$', fontsize=18)

axs[1].plot(np.linspace(0, max(bins), 1000), [scipy.stats.beta.pdf(x, 2, 2) for x in np.linspace(0, max(bins), 1000)], color = 'black')
axs[1].set_title('Analytical beta distribution', fontsize = 20)
for ax in axs:
    ax.set_xlim(left=0)
    ax.set_xlabel('$x$', fontsize=18)

plt.tight_layout()
plt.show()
```

#### Distribution used to get Bose-Einstein condensation ($\theta = 0.1$)

```python
theta = 0.1

generator.set_current_distribution('be')
generator.set_param_be(theta)

# Generate values
y_samples = []

for _ in range(n_samples):
    y_samples.append(generator.generate_value())
```

```python
# Plot the distribution
fig, axs = plt.subplots(1,2, figsize=(21, 7), sharex=True, sharey=True)
_, bins,_ = axs[0].hist(y_samples, density=True, bins = 50, width=0.01, color = 'black')
axs[0].set_title('Sampled distribution', fontsize = 20)
axs[0].set_ylabel('$p_x$', fontsize=18)

axs[1].plot(np.linspace(0, max(bins), 1000), [(1+theta)*(1-x)**theta for x in np.linspace(0, max(bins), 1000)], color = 'black')
axs[1].set_title('Analytical distribution', fontsize = 20)
for ax in axs:
    ax.set_xlim(0, 1)
    ax.set_xlabel('$x$', fontsize=18)

plt.tight_layout()
plt.show()
```
